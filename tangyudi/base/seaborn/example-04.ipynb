```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid',color_codes=True)
np.random.seed(sum(map(ord,'categorical')))
tips=sns.load_dataset('tips')
titanic=sns.load_dataset('titanic')
iris=sns.load_dataset('iris')
```

```python
titanic.head()
```

```python
sns.barplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
sns.pointplot(x='sex',y='survived',hue='class',data=titanic)
```

```python
sns.pointplot(x='class'
              ,y='survived',hue='sex'
              ,data=titanic
              ,palette={'male':'g','female':'m'}
              ,markers=["^","o"]
              ,linestyles=['-','--'])
```

```python

sns.factorplot(x='day',y='total_bill',hue='smoker',data=tips )
```

```python
sns.factorplot(x='day',y='total_bill',hue='smoker',data=tips,kind='bar')
```

```python

sns.factorplot(x='day',y='total_bill',hue='smoker',col='time',data=tips,kind='swarm')
```

```python
sns.factorplot(x='time',y='total_bill',hue='smoker',
               col='day',data=tips,kind='box',size=4,aspect=.5)
```

```python
g=sns.FacetGrid(tips,col='time')
g.map(plt.hist,'tip')
```

```python

g=sns.FacetGrid(tips,col='sex',hue='smoker')
g.map(plt.scatter,'total_bill','tip',alpha=.7)
g.add_legend()
```

```python
g=sns.FacetGrid(tips,row='smoker',col='time',margin_titles=True)
g.map(sns.regplot,'size','total_bill',color='.4',fit_reg=False,x_jitter=.1)
```

```python
g=sns.FacetGrid(tips,col='day',size=4,aspect=.5)
g.map(sns.barplot,'sex','total_bill')
```

```python
# print(tips.day)
# print(tips.day.value_counts())
# print(tips.day.value_counts().index)

ordered_days=tips.day.value_counts().index
print('1'*100)
ordered_days=[ 'Thur', 'Fri','Sat','Sun']
# print(ordered_days)

g=sns.FacetGrid(tips,row='day',row_order=ordered_days,size=1.7,aspect=4,)

g.map(sns.boxplot,'total_bill')
```

```python
from pandas import Categorical
ordered_days=tips.day.value_counts().index
print(ordered_days)
ordered_days=Categorical( [ 'Thur', 'Fri','Sat','Sun']   )
g=sns.FacetGrid(tips,row='day',row_order=ordered_days,
                size=1.7,aspect=4,)
g.map(sns.boxplot,'total_bill')
```

```python
pal=dict(Lunch='seagreen',Dinner='gray')
g=sns.FacetGrid(tips,hue='time',palette=pal,size=5)
g.map(plt.scatter,'total_bill','tip',s=50,alpha=.7,linewidth=.5,edgecolor='white')
```

```python
with sns.axes_style('white'):
    g=sns.FacetGrid(tips,row='sex',col='smoker',margin_titles=True,size=2.5)
g.map(plt.scatter,'total_bill','tip',color='#334455',edgecolor='white',lw=.5)
g.set_axis_labels('Total bill (US Dollars)','Tip')
g.set(xticks=[10,30,50],yticks=[2,6,10])
g.fig.subplots_adjust(wspace=.52,hspace=.42)
```

```python
iris=sns.load_dataset('iris')
g=sns.PairGrid(iris)
g.map(plt.scatter)
```
