```python
import numpy as np
import pandas as pd 
```

```python
rng=pd.date_range('1/1/2011',periods=90,freq='D')
```

```python
ts=pd.Series(np.random.randn(len(rng)),index=rng)
```

```python
ts.head()
```

```python
ts.resample('M').sum()
```

```python
ts.resample('3D').sum()
```

```python
ts.resample('M').sum()
```

```python
day3Ts=ts.resample('3D').mean()
```

```python
day3Ts
```

```python
day3Ts.resample('D').asfreq()
```

```python
day3Ts.resample('D').ffill(1)
```

```python
day3Ts.resample('D').interpolate()
```

```python
day3Ts.resample('D').ffill(2)
```

```python
day3Ts.resample('D').bfill()
```
